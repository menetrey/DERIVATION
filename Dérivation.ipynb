## Contexte historique

La notion de fonction dérivée est liée à la notion de fonction et de limite. Lorsque Newton(1643-1727) et Leibniz(1646-1716) créent le calcul différentiel aucune de ces deux notions sont bien définies. Un problème historique va nous permettre de d’entrevoir la difficulté de mettre en place la notion de dérivée sans avoir la notion de limite.  
**Problème:**  
Déterminer la vitesse instantané d’un objet en chute libre. On lâche une pierre à $t=0 s$. Quelle est sa vitesse instantanée au bout d’une seconde?     
Depuis Galilée, on sait que la vitesse d’une pierre en chute libre, si l’on néglige la force de frottement, **ne dépend pas de sa masse**. Galilée a pu établir alors que la position de l’objet était proportionnelle au carré du temps écoulé.

$z(t) =\frac{1}{2}gt^2=5t^2$ avec $g=10 m.s^{−2}$ . 

Pour calculer la vitesse instantanée en $t=1$, on mesure la distance entre les instants $t=1$ et $t=1+dt$, où l’intervalle de temps $dt$ est le plus petit possible (quantité infinitésimal).   

$\begin{matrix}
v(1) &=& \frac{z(1+dt)−z(1)}{dt}\\
v(1) &=& \frac{5(1+dt)^2−5}{dt}\\
v(1) &=& \frac{5+10dt+5dt^2−5}{dt} \\
v(1) &=& 10+\frac{5}{dt}\\
\end{matrix}$  

Pour Newton la vitesse en $t=1 s$ est de $10 m.s^{−1}$.  

Mais la vitesse est-elle exactement égale à $10 m.s^{−1}$ ou d’environ $10 m.s^{−1}$?    
•Si la vitesse est exactement de $10 m.s^{−1}$ alors $dt=0$   
•Mais si $dt=0$, la notion de vitesse instantanée n’a aucun sens : le dénominateur est nul.  
•Si la vitesse instantanée est d’environ $10 m.s^{−1}$, comment calculer la vitesse exacte? Ce blocage ne fut résolu qu’au $XIX^e$ siècle avec la notion de limite.    

Si cette notion de limite est cette fois rigoureuse, elle a malheureusement complexifiée le problème de départ. Avec ce nouveau concept de limite, la vitesse instantanée en $t=1$ vaut :  
$$v(1) =\lim \limits_{t\to 0}\frac{dz}{dt}$$  

La vitesse en $1$ est la limite quand $dt$ tend vers $0$ de la variation d’altitude, $dz$, sur la variation de temps $dt$.


## I) Le nombre dérivé


### 1) Définition

Le coefficient directeur $\alpha$ de la droite (AB), pour $h\neq0$, est :  
  
  
   
   
Si le point B se rapproche du point A (h tend vers 0), la droite (AB) se rapproche de la tangente (T) à la courbe en $x=a$. Le coefficient directeur de cette tangente (T) est appelé nombre dérivé noté $f′(a)$.   
  
  
  <div class="alert alert-block alert-danger">
    
   Soit la fonction $f$ définie sur un intervalle ouvert $I$ contenant $a$.<br>
    
 On appelle <b>taux d’accroissement</b> de la fonction $f$ entre $a$ et $a+h$, le nombre $t(h)$ défini par :<br> 
 $$t(h) =\frac{f(a+h)−f(a)}{h}$$
 <br>
 •La fonction $f$ admet <b>un nombre dérivé</b>, noté $f′(a)$, en $a$, si et seulement si, le taux de variation de la fonction f en $a$ admet une limite finie, c’est à dire :  
 $$f′(a) =\lim\limits_{h \to0}\frac{f(a+h)−f(a)}{h}$$
 ou encore <br>
 $$f′(a) =\lim \limits_{x \to a} \frac{f(x)−f(a)}{x−a} $$ 
 
<i>La notation $h\to 0$ signifie que $h$ tend vers zéro sans l’atteindre $(h\neq0)$.</i>  

</div>


Les physiciens utilisent la notation appelée différentielle : $f′(a) =\frac{df}{dx}(a)$.

### 3) équation de la tangente

Soit la courbe représentative d’une fonction $f$ et sa tangente en $x=a$. On a alors le schéma suivant :  



Le  coefficient  directeur  de  la  tangente  est  égal  au  nombre  dérivé  en $a$.  Si on considère un point $M(x;y)$ quelconque  de  cette  tangente,  on  obtient alors :  
$$\begin{array} \\
f′(a) & = & \frac{y−f(a)}{x−a} \\
y−f(a)& = & f′(a)(x−a) \\
y & = & f′(a)(x−a) +f(a)\\
\end{array}$$  

<div class="alert alert-block alert-danger">
    
 L’équation de la tangente en $a$ à une courbe représentative d’une fonction $f$ dérivable en $a$ est égale à :<br> 
 $$y=f′(a)(x−a) +f(a)$$
    
  </div>
  
  **Exemple:**
  <div class="alert alert-block alert-success">
    Soit la fonction $f$ définie et dérivable sur $\mathbb{R}$ par :<br>
    $$f(x) =3x^2+3x+4$$  <br>
    Déterminer l’équation de la tangente au point d’abscisse 2
    <br><br><br><br><br><br><br><br><br><br>
    </div>

```python
def fonction_mystere(a):
    for i in range(1,11):
        h = 10**-i
        t = ((a+h)**2-a**2)/h
        print(i,h,t)
    return t    
```

### Démonstration de non dérivabilité de $\sqrt{x}$ en 0.

<div class="alert alert-block alert-info">
    pour vérifier si $f(x)=\sqrt{x}$ est dérivable en 0, il faut tenter de calculer le nombre dérivée de f en 0. <br>
    calculons le taux d'accroissement:  <br>
    $$\frac{f(0+h)-f(0)}{h}= \frac{\sqrt{0+h}-\sqrt{0}}{h}= \frac{\sqrt{h}}{h}=\frac{1}{\sqrt{h}}$$
   
   Il apparaît que la limite de $\frac{1}{\sqrt{h}}$ n'est pas finie!!<br><br>
   Donc la fonction $f(x)=\sqrt{x}$ n'est pas dérivable en 0. 
 </div>

<div class="alert alert-block alert-danger">

<div class="alert alert-block alert-info">

   \end{array}$$

#### d) Dérivée de l'inverse
<div class="alert alert-block alert-danger">
    On a : $$\Big( \frac{1}{u}\Big)' = - \frac{u'}{u^2}$$
 </div>
 
 **Démonstration:**
 
 <div class="alert alert-block alert-info">
    Calculons le taux d'accroissement: <br>
    $$\begin{array}\\
    \frac{\frac{1}{u(x+h)}-\frac{1}{u(x)}}{h} & = & \frac{1}{h} \frac{u(x)-u(x+h)}{u(x)u(x+h)}\\
    \\
    & = & -\frac{1}{h}\frac{u(x+h)-u(x)}{u(x)u(x+h)}\\ \\
    & = & - \frac{\frac{u(x+h)-u(x)}{h}}{u(x)u(x+h)}
    \end{array}
    $$
    
   <br>
 On passe à la limite:<br>   
 $$\begin{array}\\
 \Big( \frac{1}{u}\Big)' & = & \lim \limits_{h \to 0}- \frac{\frac{u(x+h)-u(x)}{h}}{u(x)u(x+h)} \\ \\
 & = & \lim \limits_{h \to 0} - \frac{u(x+h)-u(x)}{h} \times \lim \limits_{h \to 0} \frac{1}{u(x)u(x+h)}\\ \\
 & = & - \frac{u'(x)}{u(x) u(x)} \\
 & = & - \frac{u'}{u^2}
 \end{array}$$
 </div>
